```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn에 내장된 데이터셋 불러오기 (mtcars는 없으므로 수동으로 생성)
url = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
df = pd.read_csv(url)

# 산점도 행렬 그리기
sns.pairplot(df,
             corner=True,           # 대각선 아래쪽만 표시
             plot_kws={'alpha': 0.7, 's': 60, 'edgecolor': 'k'},
             diag_kws={'fill': False})  # 히스토그램 스타일 옵션

plt.suptitle("Motor Trend 자동차 데이터 산점도 행렬", y=1.02)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# mtcars 데이터 불러오기
url = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
df = pd.read_csv(url)

# 상관계수 계산
corr_matrix = df.corr()

# 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, 
            annot=True,           # 숫자 표시
            cmap='coolwarm',      # 색상 스타일
            fmt=".2f",            # 소수점 자리수
            square=True, 
            linewidths=0.5, 
            cbar_kws={"shrink": 0.8})

plt.title("🚗 mtcars 데이터 변수 간 상관관계 히트맵", fontsize=16)
plt.tight_layout()
plt.show()
```
