```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# mpl.rcParams['font.family'] = 'AppleGothic'  # macOS용
mpl.rcParams['font.family'] = 'Malgun Gothic'  # Windows용
plt.rcParams['axes.unicode_minus'] = False   # 음수 부호 깨짐 방지
```

```python
titanic = sns.load_dataset('titanic')
```

```python
df = titanic[(titanic['class'] == 'First') & (titanic['alive'] == 'yes')]
```

```python
survivor_counts = df['who'].value_counts()
```

```python
colors = ['skyblue', 'lightcoral', 'gold']
plt.figure(figsize=(6, 6))
plt.pie(
    survivor_counts, 
    labels=survivor_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors,
    wedgeprops={'edgecolor': 'white'}
)
plt.title('1등급 생존자 중 성별/연령대별 비율')
plt.axis('equal')  # 원형 유지
plt.show()
```
